```python
import pandas as pd
bh=pd.read_csv('/content/HousingData.csv')
```

```python
bh.shape
```

```python
bh.isnull().sum()
```

```python
bh['CRIM']=bh['CRIM'].fillna(bh['CRIM'].mean())
```

```python
bh['ZN']=bh['ZN'].fillna(bh['ZN'].mean())
```

```python
bh['INDUS']=bh['INDUS'].fillna(bh['INDUS'].mean())
```

```python
bh['CHAS']=bh['CHAS'].fillna(bh['CHAS'].mean())
```

```python
bh['AGE']=bh['AGE'].fillna(bh['AGE'].mean())
```

```python
bh['LSTAT']=bh['LSTAT'].fillna(bh['LSTAT'].mean())
```

```python
bh.isnull().sum()
```

```python
bh.head(30)
```

```python
bh.info()
```

```python
bh['CHAS'].value_counts()
```

```python
bh=pd.get_dummies(bh,columns=['CHAS'])
```

```python
bh.head()
```

```python
x=bh.drop(['MEDV'],axis=1)
y=bh['MEDV']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import r2_score
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor(random_state=42)
model.fit(x_train,y_train)
result=model.predict(x_test)
print(r2_score(y_test,result))
mse=mean_squared_error(y_test,result)
print(np.sqrt(mse))
```

```python
import joblib as jb
jb.dump(model,"Boston housing dataset.pkl")
```
